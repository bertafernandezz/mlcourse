```python
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from feature_engine.selection import DropCorrelatedFeatures, DropDuplicateFeatures
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.model_selection import train_test_split

# Importar las clases 
from utils.transformations import MyTransformation
from utils.filters import MyFilter
```

```python
# Cargar el dataset de entrenamiento
df_train = pd.read_csv("data/preprocessed/train_data.csv")
X_train, y_train = df_train.drop(columns=['Price']), df_train['Price']
```

```python
# Crear e inicializar el preprocesador y el filtro de datos
preprocessor = MyTransformation(target_encode_cols=['city', 'Area'])  
filter = MyFilter(variance_threshold=0.01, corr_threshold=0.9, k_best=4)  
```

```python
# Ajustar el preprocesador y el filtro a los datos de entrenamiento
preprocessor.fit(X_train, y_train)
```

```python
# Transformar los datos de entrenamiento
X_processed, y_processed = preprocessor.transform(X_train, y_train)
```

```python
filter.fit(X_processed, y_processed)
```

```python
X_filtered, y_filtered = filter.transform(X_processed, y_processed)
```

```python
# Dividir los datos de entrenamiento en conjunto de entrenamiento y validacin 
X_train_final, X_val, y_train_final, y_val = train_test_split(X_filtered, y_filtered, test_size=0.2, random_state=42)
```

```python
# Entrenar el modelo XGBoost con los datos filtrados
xgboost_model = XGBRegressor(n_estimators=500, random_state=42)
xgboost_model.fit(X_train_final, y_train_final)
```

```python
xgboost_model.score(X_filtered, y_filtered)
```

```python
#  Cargar el dataset de prueba
df_test = pd.read_csv("data/preprocessed/test_data.csv")
X_test, y_test = df_test.drop(columns=['Price']), df_test['Price']
```

```python
# Preprocesar y filtrar los datos de prueba 
X_test_processed, y_test_processed = preprocessor.transform(X_test, y_test)
X_test_filtered, y_test_filtered = filter.transform(X_test_processed, y_test_processed)
```

```python
# Deshacer la transformacin para obtener las predicciones originales 
y_test_pred = xgboost_model.predict(X_test_filtered)
y_test_pred_unscaled = preprocessor.inverse_transform(y_test_pred.reshape(-1, 1))
```

```python
from sklearn.metrics import ( root_mean_squared_error, 
                             mean_absolute_error, 
                             mean_absolute_percentage_error )
```

```python
y_true = y_test.values
y_pred = y_test_pred_unscaled
rmse = root_mean_squared_error(y_true, y_pred)
mae = mean_absolute_error(y_true, y_pred)
mape = mean_absolute_percentage_error(y_true, y_pred)


metrics = {
    "RMSE ()": f"{rmse:,.2f}",
    "MAE ()": f"{mae:,.2f}",
    "MAPE (%)": f"{mape:.2%}"
}


print("\nModel Performance Metrics:\n")
print(f"{'Metric':<15} {'Value':>15}")
print("-" * 30)
for metric, value in metrics.items():
    print(f"{metric:<15} {value:>15}")
```
